```python
import wfdb
#record from wilson central terminal ecg database
record = wfdb.rdrecord('seg02', sampfrom=0,sampto=186 ,
                           channels=[4])

wfdb.plot_wfdb(record=record, title='ecg record') 
display(record.__dict__)
```

```python
from numpy import savetxt
data = record.p_signal
savetxt('data.csv',data, delimiter=',', newline=",")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
v = pd.read_csv("data.csv", header = None)
plt.subplot(2,1,1)          
plt.plot(v.iloc[0,:186])
```
